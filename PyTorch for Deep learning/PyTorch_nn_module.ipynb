```python
import torch
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__() #with the help of this the child class invokes the parent class
    #into the scene
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()
    #self.ReLU = nn.ReLU()

  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)
   # out = self.ReLU(out)

    return out
```

```python
#create dataset
features = torch.rand(10,5)

#create model
model = Model(features.shape[1])

#call model for forward pass
#model.forward(features) , we can do this too but this is not recommended

model(features)
```

```python
class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__() #with the help of this the child class invokes the parent
    # class into the scene
    self.linear1 = nn.Linear(num_features,3) #layer 1 has 3 neurons
    self.ReLU = nn.ReLU()
    self.linear2 = nn.Linear(3,1) # this is layer-2 which is getting 3 inputs
    # from the past layer (of 3 neurons), and itself has 1 neuron and hence giving
    # 1 output from this layer , thats what the 3, 1 means
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear1(features)
    out = self.ReLU(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
   # out = self.ReLU(out)

    return out
```

```python
#create dataset
features = torch.rand(10,5)

#create model
model = Model(features.shape[1])

#call model for forward pass
#model.forward(features) , we can do this too but this is not recommended

model(features)
```

```python
print (model.linear1.weight)
print (model.linear2.weight)
print (model.linear1.bias)
print (model.linear2.bias)
```

Now easy method to use the make the calss in which there is no need of lot of variables we made for each and every layer , just use the Sequential

```python
class Model(nn.Module):
  def __init__(self, num_features):

    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)

    return out
```
