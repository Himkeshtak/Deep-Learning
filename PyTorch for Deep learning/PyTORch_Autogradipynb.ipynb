Training a simple neural network.

```python
import torch

#inputs
x = torch.tensor(6.7) #input feature
y = torch.tensor(0.0) #True labels(binary)

w = torch.tensor(1.0) #Weight
b = torch.tensor(0.0) #Bias
```

```python
#binary Cross-entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8 #to prevent the log(0)
    prediction = torch.clamp(prediction, epsilon, 1-epsilon)
    return -(target * torch.log(prediction)+ (1-target)*torch.log(1-prediction))
```

```python
#Forward pass
z = w * x + b
y_pred = torch.sigmoid(z) #predicted probablity

# compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
loss
```

```python
#Derivatives
dloss_dy_pred = (y_pred - y) / (y_pred*(1-y_pred))

dy_pred_dz = y_pred * (1-y_pred)

dz_dw = x
dz_db = 1

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"dL_dw: {dL_dw}")
print(f"dL_db: {dL_db}")
```

Easy way using the auto grad to calculate the gradient descent

```python
x = torch.tensor(6.7) #input feature
y = torch.tensor(0.0) #True labels(binary
```

```python
w = torch.tensor(1.0, requires_grad = True) #Weight
b = torch.tensor(0.0, requires_grad = True) #Bias
#
```

```python
w
```

```python
b
```

```python
z = w*x +b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward() # sif ek line kaafi this , hum sirf saari equations define karni padi
```

```python
print(w.grad)
print(b.grad)
```
